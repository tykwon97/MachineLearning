```
from konlpy.tag import Mecab
import numpy as np
from gensim.models import Word2Vec, KeyedVectors

# Word2Vec 모델 로드하기
# loaded_model = KeyedVectors.load_word2vec_format("/gdrive/My Drive/colab/ko/ko.bin") -> utf-8문제 발생
loaded_model = Word2Vec.load("/gdrive/My Drive/colab/ko/ko.bin")

#한국어 Word2Vec을 이용하여 아래에 주어진 단어들을 벡터 연산한 후에 가장 가까운 상위 3개의 결과를 제시하시오.
'''
벡터 연산
1. 한국 - 서울 + 파리
2. 컴퓨터 - 과학 + 인문학
3. 사랑 + 이별
4. 충청북도 - 청주 + 강원도
5. 인간 - 사랑
6. 미술 - 물감 + 음악
'''

print(loaded_model.wv.most_similar(positive=['한국', '파리'], negative=['서울'], topn=3))
print(loaded_model.wv.most_similar(positive=['컴퓨터', '인문학'], negative=['과학'], topn=3))
print(loaded_model.wv.most_similar(positive=['사랑', '이별'], topn=3))
print(loaded_model.wv.most_similar(positive=['충청북도', '강원도'], negative=['청주'], topn=3))
print(loaded_model.wv.most_similar(positive=['인간'], negative=['사랑'], topn=3))
print(loaded_model.wv.most_similar(positive=['미술', '음악'], negative=['물감'], topn=3))

#topn : 상위 몇개의 문자를 출력할지
```
